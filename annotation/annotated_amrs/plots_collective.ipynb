```python
#%pip install rdflib
from rdflib import *
whataGraph = Graph()
whataGraph.parse("annotated_amrs.ttl", format="ttl")
```

```python

#what does a blamable eventuality involve?
cqblended = """PREFIX whata:<https://raw.githubusercontent.com/Salvadana/whata_ontology/main/ontology_development/whata_ontology.owl#> 
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX per:<http://www.ontologydesignpatterns.org/ont/persp/perspectivisation.owl#>
    PREFIX dns: <http://www.ontologydesignpatterns.org/cp/owl/description.owl#>
SELECT  ?s
WHERE {

  ?s whata:isParticipantIn ?blended.
  ?blended a whata:BlendedBlameBackground.
}
      
"""
results = whataGraph.query(cqblended)
list3=list()
for row in results:
    list3.append(row)
    print(row)
print( len (list3))
    
```

```python

#what does a blamable eventuality involve?
cqblamable = """PREFIX whata:<https://raw.githubusercontent.com/Salvadana/whata_ontology/main/ontology_development/whata_ontology.owl#> 
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX per:<http://www.ontologydesignpatterns.org/ont/persp/perspectivisation.owl#>
    PREFIX dns: <http://www.ontologydesignpatterns.org/cp/owl/description.owl#>
SELECT  ?s 
WHERE {
 
  ?s whata:isParticipantIn ?b.
  
  ?b a whata:Blamable.
}
      
"""
results = whataGraph.query(cqblamable)
list1=list()
for row in results:
    list1.append(row)
    print(row)
print(len(list1))
```

```python

#what does a blamable eventuality involve?
cqc_blamable = """PREFIX whata:<https://raw.githubusercontent.com/Salvadana/whata_ontology/main/ontology_development/whata_ontology.owl#> 
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX per:<http://www.ontologydesignpatterns.org/ont/persp/perspectivisation.owl#>
    PREFIX dns: <http://www.ontologydesignpatterns.org/cp/owl/description.owl#>
SELECT  ?s
WHERE {

  ?s whata:isParticipantIn ?b.
  ?b a whata:CounterBlamable.
}
      
"""
results = whataGraph.query(cqc_blamable)
list2=list()
for row in results:
    list2.append(row)
print(len(list2))
```

```python
set1=set(list1)
set2=set(list2)
set3=set(list3)
print (set1, set2, set3)
```

```python
# Function to extract the part after the last #
def extract_label(uri):
    return str(uri).split('#')[-1]

# Extract labels
set1_labels = {extract_label(uri[0]) for uri in set1}
set2_labels = {extract_label(uri[0]) for uri in set2}
set3_labels = {extract_label(uri[0]) for uri in set3}
```

```python
print(set1_labels,set2_labels,set3_labels)
```

```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn3



# Increase the dimensions of the plot
plt.figure(figsize=(10, 10))

# Create a Venn diagram
venn_diagram = venn3([set1_labels, set2_labels, set3_labels], set_labels=('Blamable', 'CounterBlamable', 'Blended'))


# Set colors for each subset

venn_diagram.get_patch_by_id('001').set_color('green')

venn_diagram.get_patch_by_id('111').set_color('gray')


# Customize the labels to show the elements instead of counts
if venn_diagram.get_label_by_id('100'):
    venn_diagram.get_label_by_id('100').set_text(',\n '.join(map(str, set1_labels - set2_labels - set3_labels)))
if venn_diagram.get_label_by_id('010'):
    venn_diagram.get_label_by_id('010').set_text(',\n '.join(map(str, set2_labels - set1_labels - set3_labels)))
if venn_diagram.get_label_by_id('001'):
    venn_diagram.get_label_by_id('001').set_text(',\n '.join(map(str, set3_labels - set1_labels - set2_labels)))
if venn_diagram.get_label_by_id('110'):
    venn_diagram.get_label_by_id('110').set_text(',\n '.join(map(str, set1_labels & set2_labels - set3_labels)))
if venn_diagram.get_label_by_id('101'):
    venn_diagram.get_label_by_id('101').set_text(',\n '.join(map(str, set1_labels & set3_labels - set2_labels)))
if venn_diagram.get_label_by_id('011'):
    venn_diagram.get_label_by_id('011').set_text(',\n '.join(map(str, set2_labels & set3_labels - set1_labels)))
if venn_diagram.get_label_by_id('111'):
    venn_diagram.get_label_by_id('111').set_text(',\n '.join(map(str, set1_labels & set2_labels & set3_labels)))

# Show the plot
plt.title("Venn Diagram of Responsibility Owner instances in the Blamable, CounterBlamable and Blended Space")
plt.show()
```
