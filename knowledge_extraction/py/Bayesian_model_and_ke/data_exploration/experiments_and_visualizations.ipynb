```python
import nltk

nltk.download('averaged_perceptron_tagger')
nltk.download('popular')
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
```

```python


# Read text from corpus files
with open('../txt_for_classification/result_whatabout.txt', 'r', encoding='utf-8') as file:
    corpus1 = file.read()
    
with open('../txt_for_classification/no_what_no_semeval.txt', 'r' , encoding='utf-8') as file:
    corpus2 = file.read()

# Tokenize text and count pronouns for corpus 1
tokens1 = word_tokenize(corpus1)
pronouns1 = [word.lower() for word, pos in nltk.pos_tag(tokens1) if pos == 'PRP']
pronoun_freq1 = nltk.FreqDist(pronouns1)

# Tokenize text and count pronouns for corpus 2
tokens2 = word_tokenize(corpus2)
pronouns2 = [word.lower() for word, pos in nltk.pos_tag(tokens2) if pos == 'PRP']
pronoun_freq2 = nltk.FreqDist(pronouns2)

# Plotting
plt.bar(pronoun_freq1.keys(), pronoun_freq1.values(), color='blue', alpha=0.5, label='Corpus 1')
plt.bar(pronoun_freq2.keys(), pronoun_freq2.values(), color='red', alpha=0.5, label='Corpus 2')
plt.xlabel('Pronouns')
plt.ylabel('Frequency')
plt.title('Frequency of Pronouns in Corpora')
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
pronouns = list(pronoun_freq1.keys())  # Assuming both corpora have the same pronouns
frequency_corpus1 = list(pronoun_freq1.values())
frequency_corpus2 = list(pronoun_freq2.values())

# Set the width of the bars
bar_width = 0.35

# Set the x locations for the groups
index = np.arange(len(pronouns))

# Plot bars
plt.figure(figsize=(15, 6))
plt.bar(index, frequency_corpus1, bar_width, color='blue', label='Corpus 1')
plt.bar(index + bar_width, frequency_corpus2, bar_width, color='red', label='Corpus 2')

# Add labels and title
plt.xlabel('Pronouns')
plt.ylabel('Frequency')
plt.title('Frequency of Pronouns in Corpora')
plt.xticks(index + bar_width / 2, pronouns)  # Centering labels on bars
plt.legend()

# Show plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
