```python
import spacy

# Load the English language model
nlp = spacy.load("en_core_web_sm")

# Example sentences
sentences = [
    "The cat chases the mouse.",
    "The dog barks loudly.",
    "The bird flies in the sky."
]

# Perform analysis for each sentence
for sentence in sentences:
    doc = nlp(sentence)
    
    # Syntactic structure analysis
    noun_phrases = [chunk.text for chunk in doc.noun_chunks]
    verb_phrase = [token.text for token in doc if token.dep_ == "ROOT"]
    
    print("Sentence:", sentence)
    print("Noun Phrases:", noun_phrases)
    print("Verb Phrase:", " ".join(verb_phrase))
    
    # Dependency structure analysis
    for token in doc:
        print(token.text, "-->", token.dep_, ":", token.head.text)
    
    print()
```

```python
from nltk.corpus import wordnet as wn

# Example text
text = "The cat chases the mouse."

# Tokenize the text and get WordNet synsets for each word
tokens = text.lower().split()
synsets = {word: wn.synsets(word) for word in tokens}

# Extract concepts and relations
concepts = set()
relations = set()

for word, syns in synsets.items():
    # Extract concepts
    for syn in syns:
        concepts.add(syn.name().split('.')[0])
        
    # Extract relations
    for hypernym in syn.hypernyms():
        relations.add((word, "is_a", hypernym.lemma_names()[0]))

# Print concepts and relations
print("Concepts:", concepts)
print("Relations:", relations)
```
